```python
import cv2
import os

def extract_frames_from_videos(vf, of, interval=30):
    if not os.path.exists(of):
        os.makedirs(of)
    for v in os.listdir(vf):
        if v.endswith('.mp4'):
            v_path = os.path.join(vf, v)
            cap = cv2.VideoCapture(v_path)
            index = 0
            while True:
                ret, frame = cap.read()
                if not ret:
                    break
                if index % interval == 0:
                    frame_filename = f"{os.path.splitext(v)[0]}_frame_{index}.jpg"
                    frame_path = os.path.join(of, frame_filename)
                    cv2.imwrite(frame_path, frame)
                index += 1
            cap.release()

# Call the function
extract_frames_from_videos('/content/mainset/_dataset/real', '/content/mainset/_dataset/frames_real')
extract_frames_from_videos('/content/mainset/_dataset/fake', '/content/mainset/_dataset/frames_fake')
```

```python
import os

def count_items_in_folder(folder_path):
  num_items = len(os.listdir(folder_path))
  return num_items
print("Real : ",count_items_in_folder('/content/mainset/_dataset/real'))
print("Fake: ",count_items_in_folder('/content/mainset/_dataset/fake'))
print("Frames_Real : ",count_items_in_folder('/content/mainset/_dataset/frames_real'))
print("Frames_Fake: ",count_items_in_folder('/content/mainset/_dataset/frames_fake'))
```

```python
!pip install mtcnn
from mtcnn import MTCNN
import cv2
import os

detector = MTCNN()

def detect_faces(inputf, outputf):
    if not os.path.exists(outputf):
        os.makedirs(inputf)

    for fname in os.listdir(inputf):
        image_path = os.path.join(inputf, fname)
        image = cv2.imread(image_path)

        result = detector.detect_faces(image)

        if result:
            x, y, width, height = result[0]['box']
            face = image[y:y+height, x:x+width]

            cv2.imwrite(os.path.join(outputf, fname), face)

detect_faces('/content/mainset/_dataset/frames_real', '/content/mainset/_dataset/faces_real')
detect_faces('/content/mainset/_dataset/frames_fake', '/content/mainset/_dataset/faces_fake')
```

```python
from PIL import Image
import os

def resize_images(input_folder, output_folder, size=(224, 224)):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        img_path = os.path.join(input_folder, filename)
        img = Image.open(img_path)
        img_resized = img.resize(size)
        img_resized.save(os.path.join(output_folder, filename))

resize_images('/content/mainset/_dataset/faces_real', '/content/mainset/_dataset/resized_real', (224, 224))
resize_images('/content/mainset/_dataset/faces_fake', '/content/mainset/_dataset/resized_fake', (224, 224))
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rescale=1.0/255.0,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)
train_data = datagen.flow_from_directory(
    '/content/mainset/_dataset/resized/',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='training'
)
val_data = datagen.flow_from_directory(
    '/content/mainset/_dataset/resized/',
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

def build_cnn_model():
    Deepfake_Detection_model = Sequential()

    # First convolutional layer
    Deepfake_Detection_model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)))
    Deepfake_Detection_model.add(MaxPooling2D((2, 2)))

    # Second convolutional layer
    Deepfake_Detection_model.add(Conv2D(64, (3, 3), activation='relu'))
    Deepfake_Detection_model.add(MaxPooling2D((2, 2)))

    # Flatten output
    Deepfake_Detection_model.add(Flatten())

    # Fully connected layers
    Deepfake_Detection_model.add(Dense(128, activation='relu'))
    Deepfake_Detection_model.add(Dropout(0.5))
    Deepfake_Detection_model.add(Dense(64, activation='relu'))
    Deepfake_Detection_model.add(Dense(1, activation='sigmoid'))  # Binary classification (real vs fake)

    Deepfake_Detection_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    return Deepfake_Detection_model

# Build the model
Deepfake_Detection_model = build_cnn_model()

# Print model summary
Deepfake_Detection_model.summary()
```

```python
history = Deepfake_Detection_model.fit(
    train_data,
    validation_data=val_data,
    epochs=20,
    verbose=1
)
```

```python
loss, accuracy = Deepfake_Detection_model.evaluate(val_data)
print(f'Test Accuracy: {accuracy * 100:.2f}%')
```
